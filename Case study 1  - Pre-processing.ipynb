```python
#Preprocessing data
import pandas as pd
```

```python
raw_data = pd.read_csv("Absenteeism-data.csv")
type(raw_data)
```

```python
raw_data.head(5)
```

```python
df1 = raw_data.copy()
df1.info()
```

```python
#dropping unnecessary columns
df1 = df1.drop(['ID'], axis = 1)
```

```python
df1['Reason for Absence'].min()
```

```python
df1['Reason for Absence'].max()
```

```python
sorted(df1['Reason for Absence'].unique())
```

```python
#Get dummies for reasons for Absense column
```

```python
reason_columns = pd.get_dummies(df1['Reason for Absence'], drop_first = True)
reason_columns
```

```python
#Grouping reasons of absense
```

```python
df1 = df1.drop(['Reason for Absence'], axis = 1)
```

```python
reasontype_1 = reason_columns.loc[:, 1:14].max(axis=1)
reasontype_2 = reason_columns.loc[:, 15:17].max(axis=1)
reasontype_3 = reason_columns.loc[:, 18:21].max(axis=1)
reasontype_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
#Attaching these columns to the original dataset
```

```python
df1 = pd.concat([df1, reasontype_1, reasontype_2, reasontype_3, reasontype_4], axis = 1)
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
df1.columns = column_names
```

```python
#Re-ordering the columns
```

```python
reordered_cols = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df1 = df1[reordered_cols]
```

```python
df2 = df1.copy()
```

```python
#Making changes to the Date coulumn
```

```python
df2['Date'] = pd.to_datetime(df2['Date'])
df2['Date']
```

```python
df2['Date'] = pd.to_datetime(df2['Date'], format = '%d/%m/%Y')
```

```python
#Extracting the month
```

```python
list_months = []
for i in range(df2.shape[0]):
    list_months.append(df2['Date'][i].month)
df2['Month Value'] = list_months #adding month column to the data
```

```python
#Extracting the weekday
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
df2['Day of the Week'] = df2['Date'].apply(date_to_weekday) #adding weekday column to the data
```

```python
df2 = df2.drop(['Date'], axis = 1)
df2.head(5)
```

```python
#changing column names
column_names = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
df2 = df2[column_names]
df2.head(5) 
```

```python
df3 = df2.copy()
```

```python
#Making changes to the Edcuation column
```

```python
df3['Education'].value_counts()
```

```python
#Segregating into higher and lower education
df3['Education'] = df3['Education'].map({1:0, 2:1, 3:1, 4:1})
df3
```

```python
#Checkpoint
```

```python
preprocessed_data = df3.copy()
```

```python
preprocessed_data.to_csv('preprocessed_data.csv', index = False)
```
